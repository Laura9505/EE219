```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, KNNWithMeans, accuracy
from surprise.model_selection import cross_validate, KFold, train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
%matplotlib inline

dataset = pd.read_csv('ratings.csv', usecols=['userId', 'movieId', 'rating'])
#R = dataset.pivot_table(index=['userId'], columns=['movieId'], values='rating', fill_value=0)
#print (R)
#print (R.shape)

#question 10
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(dataset[['userId', 'movieId', 'rating']], reader)

sim_options = {'name': 'pearson'}
algo = KNNWithMeans(sim_options = sim_options)
#report = cross_validate(algo, data, measures = ['RMSE', 'MAE'], cv = 10, verbose = True)

def plot_rmse(error):
	plt.plot(range(2,102,2), error)
	plt.title('average RMSE against k')
	plt.ylabel('average rmse')
	plt.xlabel('k')
	plt.show()

def plot_mae(error):
	plt.plot(range(2,102,2), error)
	plt.title('average MAE against k')
	plt.ylabel('average mae')
	plt.xlabel('k')
	plt.show()

k_value = np.arange(2, 101, 2)
mean_mae = []
mean_rmse = []

for k in k_value:
	algo = KNNWithMeans(k = k, sim_options = sim_options)
	print ("k=%d :" % k)
	report = cross_validate(algo, data, measures = ['RMSE', 'MAE'], cv = 10, verbose = True)
	mean_mae.append(np.mean(report['test_mae']))
	mean_rmse.append(np.mean(report['test_rmse']))

plot_rmse(mean_rmse)
plot_mae(mean_mae)

#question 12 -- popular movie trimming
kf = KFold(n_splits = 10)
k_value = np.arange(2, 101, 2)
mean_rmse_pop = []

def trim_pop(subset):
	test = pd.DataFrame(subset, columns=['userId', 'movieId', 'rating'])
	testset = test.groupby("movieId").filter(lambda x: len(x) > 2)
	testdata = np.array(testset)
	testset_trim = testdata.tolist()
	return testset_trim

for k in k_value:
	print ('k = %d :' %k)
	algo = KNNWithMeans(k = k, sim_options = sim_options)
	rmse_val = []
	for trainset, testset in kf.split(data):
		algo.fit(trainset)
		testset = trim_pop(testset)
		predictions = algo.test(testset)
		rmse_val.append(accuracy.rmse(predictions))
		print (rmse_val)
	mean_rmse_pop.append(np.mean(rmse_val))

print ('The minimum average RMSE: ')
print (np.min(mean_rmse_pop))
plot_rmse(mean_rmse_pop)

#question 13 -- unpopular movie trimming
kf = KFold(n_splits = 10)
k_value = np.arange(2, 101, 2)
mean_rmse_unpop = []

def trim_unpop(subset):
	test = pd.DataFrame(subset, columns=['userId', 'movieId', 'rating'])
	testset = test.groupby("movieId").filter(lambda x: len(x) <= 2)
	testdata = np.array(testset)
	testset_trim = testdata.tolist()
	return testset_trim

for k in k_value:
	print ('k = %d :' %k)
	algo = KNNWithMeans(k = k, sim_options = sim_options)
	rmse_val = []
	for trainset, testset in kf.split(data):
		algo.fit(trainset)
		testset = trim_unpop(testset)
		predictions = algo.test(testset)
		rmse_val.append(accuracy.rmse(predictions))
		print (rmse_val)
	mean_rmse_unpop.append(np.mean(rmse_val))
	
print ('The minimum average RMSE: ')
print (np.min(mean_rmse_unpop))
plot_rmse(mean_rmse_unpop)

#question 14 -- high variance movie trimming
kf = KFold(n_splits = 10)
k_value = np.arange(2, 101, 2)
mean_rmse_highvar = []

def trim_highvar(subset):
	test = pd.DataFrame(subset, columns=['userId', 'movieId', 'rating'])
	testset = test.groupby("movieId").filter(lambda x: np.var(x['rating']) >= 2 and len(x) > 5)
	testdata = np.array(testset)
	testset_trim = testdata.tolist()
	return testset_trim

for k in k_value:
	print ('k = %d :' %k)
	algo = KNNWithMeans(k = k, sim_options = sim_options)
	rmse_val = []
	for trainset, testset in kf.split(data):
		algo.fit(trainset)
		testset = trim_highvar(testset)
		predictions = algo.test(testset)
		rmse_val.append(accuracy.rmse(predictions))
		print (rmse_val)
	mean_rmse_highvar.append(np.mean(rmse_val))
	
print ('The minimum average RMSE: ')
print (np.min(mean_rmse_highvar))
plot_rmse(mean_rmse_highvar)

#question 15
from surprise.model_selection import cross_validate, KFold, train_test_split
from sklearn.metrics import roc_curve, auc

threshold_val = [2.5, 3, 3.5, 4]

def plot_roc_curve(y_test, y_score, threshold):
    fpr, tpr, thresholds = roc_curve(y_test, y_score)
    fig, ax = plt.subplots()
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, lw = 2, label = 'area under curve = %0.4f' % roc_auc)
    ax.grid(color = '0.7', linestyle = '--', linewidth = 1)
    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontsize = 15)
    ax.set_ylabel('True Positive Rate', fontsize = 15)
    plt.title('ROC Curve (threshold = %f)' % threshold)
    ax.legend(loc = "lower right")
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontsize(15)
    plt.show()

for threshold in threshold_val:

    data = Dataset.load_from_df(dataset[['userId', 'movieId', 'rating']], reader)
    trainset, testset = train_test_split(data, test_size=0.1)

    label_true_test = []
    for i in range(len(testset)):
        if testset[i][2] >= threshold:
            label_true_test.append(1)
        else:
            label_true_test.append(0)

    algo = KNNWithMeans(k = 30, sim_options = sim_options)
    algo.fit(trainset)
    predictions = algo.test(testset)

    prediction = []
    for i in range(len(predictions)):
        prediction.append(predictions[i][3])

    plot_roc_curve(label_true_test, prediction, threshold)
```
